# Problema 1

Cerinte:

   a) Permutari cuvant 'word'

        itertools.permutations(p,r) = permutari a lui p de lungime r
        itertools.combinations(p,r) = combinari a lui p de lungime r

   b) Numarul total al permutarilor

        Functia math.perm(n,k) = aranjamente de n luate cate k (k == n rezulta permutari)
        math.comb(n,k) = combinari

   c) O permutare aleatoare

        random.randint(0, n) = numar random din [0,n]

```python
import itertools
from math import perm, comb
from itertools import combinations, permutations
from random import randint

print(list(permutations('word')))
print(perm(len('word'), len('word')))
print(list(permutations('word', 4))[randint(0, len(list(permutations('word', 4))))])
```

# Problema 2

Problema 1 doare pt combinari si aranjamente


```python
def get_combinations(word, k):
    return (list(combinations(word, k)),
            comb(len(word),k),
            list(combinations(word,k))[randint(0, len(list(combinations(word, k))))])
def get_arangements(word, k):
    return (list(permutations(word, k)),
            perm(len(word),k),
            list(permutations(word,k))[randint(0, len(list(permutations(word, k))))])

print(get_combinations('word', 3))
print(get_arangements('word', 3))
```

# Problema 3

Toate combinarile cu repetitie de k elemnte dintr-un grup de litere
    combinations_with_replacement(p,r) = toate combinarile cu repetitie de lungime r


```python
from itertools import combinations_with_replacement
def get_combs_repetition(word, k):
    return list(combinations_with_replacement(word, k))

print(get_combs_repetition('ABCDE', 4))
```

# Problema 4

In cate moduri se pot aseza 5 persoane pe 12 scaune astfel incat intre ele sa existe cel putin un loc liber

```python
def nr_moduri(nr_persoane, scaune):
    locuri_libere = scaune - nr_persoane + 1
    return perm(locuri_libere, nr_persoane)
print(nr_moduri(5, 12))
```
