# Problema 1

```python
from scipy.stats import expon,randint
from matplotlib import pyplot as plt
from numpy import floor,linspace,mean
alpha = 1/2
n = 10

T = expon.rvs(loc=0, scale=1/alpha, size=10000)
bins = 10
interval = linspace(0,10,1000)
plt.hist(T, bins=bins,range=(0,10),density=True,edgecolor='black',label='frecvente relative',color='blue',alpha=0.8)
plt.plot(interval, expon.pdf(interval,loc=0,scale=1/alpha), label='functie de densitate',color='red')

plt.xticks(range(0,11,2))
plt.grid()
plt.legend()
plt.show()

prob_empirica_N_mai_mare_ca_5 = mean([randint.rvs(1,11)-floor(t)>=5 for t in T])
prob_teoretica = sum([randint.pmf(k,1,11)*expon.cdf(k-4,scale=1/alpha) for k in range(5,11)])

print(f"Probabilitatea din simulari ca nota sa fie >= 5 este: {prob_empirica_N_mai_mare_ca_5:.4f}")
print(f"Probabilitatea teoretica ca nota sa fie >= 5 este: {prob_teoretica:.4f}")
```

# Problema 2

```python
from scipy.stats import geom,norm

persoane = norm.rvs(loc=1.65,scale=0.2, size=10000)
prob_succes = norm.sf(1.90,loc=1.65,scale=0.2)

X = geom.rvs(prob_succes,size=10000)
plt.hist(X,bins=[k+0.5 for k in range (1,11)],density=True,label='frecvente relative',edgecolor='k')
plt.xticks(range(1,11))
pX_mai_mare_10 = mean([incercare > 10 for incercare in X])
pX_mai_mare_10_teoretica = geom.sf(10,prob_succes)

print(f"Probabilitate estimata: {pX_mai_mare_10:.4f}")
print(f"Probabilitate teoretica: {pX_mai_mare_10_teoretica:.4f}")

plt.legend()
plt.grid()
plt.show()

```
